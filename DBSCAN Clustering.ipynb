```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Dataset

```python
from sklearn.datasets import make_circles, make_moons
n_samples = 2000

circles_dataset = make_circles(n_samples=n_samples, noise=0.05, factor=0.5, random_state=170)
moons_dataset = make_moons(n_samples=n_samples, noise=0.05, random_state=170)
```

```python
moons_x = moons_dataset[0]
moons_y = moons_dataset[1]

circles_x = circles_dataset[0]
circles_y = circles_dataset[1]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 6))
axes[0].scatter(circles_x[:, 0], circles_x[:, 1])
axes[1].scatter(moons_x[:, 0], moons_x[:, 1])
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan_moons = DBSCAN(eps=0.05, min_samples=6)
dbscan_moons.fit(moons_x)
dbscan_moons_labels = dbscan_moons.labels_
```

```python
np.unique(dbscan_moons_labels)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 6))
axes[0].scatter(moons_x[:, 0], moons_x[:, 1], c=moons_y)
axes[1].scatter(moons_x[:, 0], moons_x[:, 1], c=dbscan_moons_labels)
```

```python
dbscan_circles = DBSCAN(eps=0.06, min_samples=6)
dbscan_circles.fit(circles_x)
dbscan_circles_labels = dbscan_circles.labels_
```

```python
np.unique(dbscan_circles_labels)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 6))
axes[0].scatter(circles_x[:, 0], circles_x[:, 1], c=circles_y)
axes[1].scatter(circles_x[:, 0], circles_x[:, 1], c=dbscan_circles_labels)
```
